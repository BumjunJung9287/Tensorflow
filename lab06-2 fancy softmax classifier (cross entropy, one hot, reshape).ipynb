```python
import tensorflow as tf
import numpy as np
xy = np.loadtxt("data-04-zoo.csv",delimiter=",",dtype=np.float32)
x_data = xy[:,0:-1]
y_data = xy[:,[-1]]
```

```python
nb_classes = 7 #0~6

X = tf.placeholder(tf.float32, [None,16])
Y = tf.placeholder(tf.int32,[None,1]) # 0~6 shape=(?,1)  ex)[[0],[3]]

Y_one_hot = tf.one_hot(Y,nb_classes) # one hot shape=(?,1,7)   ex)[[0],[3]] -> [[[10000000],[0001000]]]
Y_one_hot = tf.reshape(Y_one_hot, [-1,nb_classes]) #shape=(?,7) #one_hot에서 더해준 차원하나를 뺴줌
# ex) [[10000000],[0001000]]  <- reshaped

W = tf.Variable(tf.random_normal([16,nb_classes],name = "weight"))
b = tf.Variable(tf.random_normal([nb_classes]),name="bias")

#logit = score (softmax집어넣기전 y값)
logits = tf.matmul(X,W)+b
hypothesis = tf.nn.softmax(logits)
#cost = tf.reduce_mean(-tf.reduce_sum(Y*tf.log(hypothesis),axis = 1))
cost_i = tf.nn.softmax_cross_entropy_with_logits(logits = logits,labels = Y_one_hot)
cost = tf.reduce_mean(cost_i) #동일한 cost

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(cost)

prediction = tf.argmax(hypothesis,1) ##예측한 확률을 0~6사이의 값중 하나로 만들어냄 (argmax를 예측값으로 생각)
correct_prediction = tf.equal(prediction,tf.argmax(Y_one_hot,1)) #Y 그자체 (one_hot만들기전의 값)
accuracy = tf.reduce_mean(tf.cast(correct_prediction,tf.float32))


with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(2000):
        sess.run(optimizer,feed_dict = {X : x_data, Y: y_data})
        if step%200==0:
            loss,acc = sess.run([cost,accuracy],feed_dict = {X : x_data, Y: y_data})
            print("Step : {:5}\tloss: {:.3f}\tAcc: {:.2%}".format(step,loss,acc))
      
    #Let's see if we can predict
    pred = sess.run(prediction,feed_dict = {X:x_data})
    
    #y_data : (N,1) = flatten => (N,) matches pred.shape
    for p,y in zip(pred,y_data.flatten()):
        print("[{}] Prediction : {}   True Y: {}".format(p==int(y),p,int(y)))
    
```
