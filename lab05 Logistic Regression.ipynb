```python
import tensorflow as tf
x_data = [[1,2], [2,3], [3,1], [4,3], [5,3], [6,2]]
y_data = [[0],[0],[0],[1],[1],[1]]

X = tf.placeholder(tf.float32, shape = [None, 2])
Y = tf.placeholder(tf.float32, shape = [None, 1])
W = tf.Variable(tf.random_normal([2,1]), name = "weight")
b = tf.Variable(tf.random_normal([1], name = "bias"))  ## dot(X,W) + b 식에서 형태 유추

# Hy using sigmoid: tf.div(1., 1+tf.exp(tf.matmul(X, W) + b)) 과 동일
hypothesis = tf.sigmoid(tf.matmul(X, W) + b)

#cost/loss function
cost = -tf.reduce_mean(Y * tf.log(hypothesis) + (1-Y) * \
                      tf.log(1 - hypothesis))

optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.01)
train = optimizer.minimize(cost)

#Accuacy computation
#True if hy > 0.5 else False ## hy값이 0.5보다 크면 1로 추정, 나머지 0
predicted = tf.cast(hypothesis > 0.5, dtype = tf.float32) #True -> 1, False -> 0

# equal이면 1반환 (1이면 100%) 
## tf.cast(조건,dtype = tf.float32) : 조건이 참이면 1, 거짓이면 0 반환.
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype = tf.float32))

with tf.Session() as sess:
    #initialize TensorFlow variables
    sess.run(tf.global_variables_initializer())
    
    for step in range(10001):
        cost_val, _ = sess.run([cost, train], feed_dict = {X : x_data, Y : y_data})
        #if step%200 == 0:
        #    print(step,"cost : ", cost_val)
    #Accuract report
    h, c, a = sess.run([hypothesis, predicted, accuracy],feed_dict = {X : x_data, Y : y_data})
    print("Hypothesis : {}\npredicted : {}, \naccuracy : {}".format(h,c,a))
```

```python
#Logistic Regression 혼자 짜보기! 도전
import tensorflow as tf
x_data = [[1,2], [2,3], [3,1], [4,3], [5,3], [6,2]]
y_data = [[0],[0],[0],[1],[1],[1]]

X = tf.placeholder(shape = [None, 2], dtype = tf.float32)
Y = tf.placeholder(tf.float32, shape = [None, 1])
W = tf.Variable(tf.random_normal([2,1]), name = "weight")
b = tf.Variable(tf.random_normal([1]), name = "bias")

hy = tf.sigmoid(tf.matmul(X,W) + b)
# hy = div(1., 1 + tf.exp(tf.matmul(X,W) + b))
cost = -tf.reduce_mean(Y*tf.log(hy) + (1-Y)*tf.log(1 - hy))

optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.01)
train = optimizer.minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())

predicted = tf.cast(hy > 0.5, dtype =tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype = tf.float32))
for step in range(10001):
    cost_val, hy_val, _ = sess.run([cost, hy, train], feed_dict = {X : x_data, Y : y_data})
    if step %2000 == 0:
        print(step,cost_val,hy_val)
        
h, p, a = sess.run([hy, predicted, accuracy], feed_dict = {X:x_data, Y: y_data})
print(h,"\n",p,"\n",a)
```

```python
##logistic 복습.
import tensorflow as tf
x_data = [[1,2], [2,3], [3,1], [4,3], [5,3], [6,2]]
y_data = [[0],[0],[0],[1],[1],[1]]

X = tf.placeholder(tf.float32,shape = [None,2])
Y = tf.placeholder(tf.float32,shape = [None,1])
W = tf.Variable(tf.random_normal([2,1]),name = "Weight")
b = tf.Variable(tf.random_normal([1]),name = "bias")

Hypothesis = tf.sigmoid(tf.matmul(X,W)+b)

cost = -tf.reduce_mean(Y*tf.log(Hypothesis)+(1-Y)*tf.log(1-Hypothesis))

optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.01)
train = optimizer.minimize(cost)

sess = tf.Session()

sess.run(tf.global_variables_initializer())

Predicted = tf.cast(Hypothesis > 0.5,dtype = tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(Predicted,Y),dtype = tf.float32))

for step in range(10000):
    #sess.run(train)
    cost_val, hy_val, _ = sess.run([cost,Hypothesis,train], feed_dict = {X:x_data, Y:y_data})
    if step%2000 == 0:
        print(step)
        print(cost_val,"\n", hy_val)
        
hy,p,a = sess.run([Hypothesis,Predicted,accuracy], feed_dict = {X:x_data,Y:y_data})
print(h,p,a)
```
