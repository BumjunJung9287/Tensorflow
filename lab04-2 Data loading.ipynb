```python
import numpy as np

xy = np.loadtxt('sample_test_score01.csv', delimiter =',', dtype = np.float32)
x_data = xy[:, 0:-1] #처음 슬라이스는 행 다음슬라이스는 열을 의미하는듯 (모든 행을 가져온 것.)
y_data = xy[:, -1:]

print(x_data.shape, x_data, len(x_data))
print(y_data.shape, y_data)
```

```python
import tensorflow as tf

X = tf.placeholder(tf.float32, shape = [None, 3])
Y = tf.placeholder(tf.float32, shape = [None, 1])
W = tf.Variable(tf.random_normal(shape = [3,1]), name = "weight")
b = tf.Variable(tf.random_normal(shape = [1]), name = "bias")

hy = tf.matmul(X,W) + b
cost = tf.reduce_mean(tf.square(hy - Y))

optimizer = tf.train.GradientDescentOptimizer(learning_rate = 1e-5)
train = optimizer.minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(2001):
    cost_val, hy_val, _ = sess.run([cost, hy, train], feed_dict = {X : x_data, Y: y_data})
    if step %100 == 0:
        print(step, cost_val, hy_val)
```
